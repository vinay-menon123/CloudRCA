```python
df = pd.read_csv("./data/test_2022-04-26_07_59_40.csv")
words = df["words"]
len(words)
```

```python
results = []
```

```python
flag = 0
for i in words:
    l2 = []
    for j in danger_words:
        if j in i:
            l2.append(j)
            flag=1
    if(flag==0):
        results.append("")
    if(flag==1):
        results.append(l2)
        flag=0
```
